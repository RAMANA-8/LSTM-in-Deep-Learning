```python
import numpy as np #numpy
from tensorflow.keras.models import Sequential #keras layer - linear stack
from tensorflow.keras.layers import LSTM, Dense #outputting the final prediction

# Generate dummy sequential data
data = np.array([[i + j for j in range(3)] for i in range(1000)]) #[0,1,2],[1,2,3]...[997,998,999]
labels = np.array([i + 3 for i in range(1000)]) #next number if [0,1,2]->3

# Reshape data to fit LSTM (samples, time steps, features)
data = data.reshape((1000, 3, 1)) # conveting 2D into 3D

# Split data into training and test sets
train_size = int(len(data) * 0.8) #80%
train_data, test_data = data[:train_size], data[train_size:]# fisrt 800 elements
train_labels, test_labels = labels[:train_size], labels[train_size:] # remain 200 elements

# Build a simple LSTM model
model = Sequential() #model
model.add(LSTM(50, activation='relu', input_shape=(3, 1))) #add-input gate, 50- memory cell, Activation function- forget gate
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(train_data, train_labels, epochs=200, validation_data=(test_data, test_labels))

# Make predictions
predictions = model.predict(test_data)

# Print a few predictions
for i in range(5):
    print(f"Expected: {test_labels[i]}, Predicted: {predictions[i][0]}")
```
